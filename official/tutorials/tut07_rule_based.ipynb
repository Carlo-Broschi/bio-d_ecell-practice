分子に**サイト（結合手）と状態**を持たせ、ワイルドカード `^_` や名前付きワイルドカード `_1` で
「多くの具体反応をまとめた1本のルール」を書く。組合せ爆発する重合・翻訳後修飾を扱える。

```python
import warnings; warnings.filterwarnings('ignore')
from ecell4.prelude import *

# サイト付き種の例
print('bonded pair :', Species('A(b^1).B(b^1)').serial())
print('site state  :', Species('A(b=u)').serial())

# 重合ルール: A の右手 r と別の A の左手 l が結合
with reaction_rules():
    A(r) + A(l) > A(r^1).A(l^1) | 1.0
m1 = get_model(is_netfree=True)      # ルールのまま保持
print('netfree rules:', len(m1.reaction_rules()))

# 具体反応へ展開（各分子 A を最大 4 量体までに制限）
m2 = m1.expand([Species('A(l, r)')], 100, {Species('A'): 4})
print('expanded: species=%d, reactions=%d' % (len(m2.list_species()), len(m2.reaction_rules())))
```

**要点**:
- `A(b^1).B(b^1)` = A と B が手 `b` 同士で結合（`^数字`が結合、`.`が複合体）。`A(b=u)` = 手 `b` が状態 `u`。
- ワイルドカード `A(p=u^_)` = 「p が u 状態で何かに結合済み」。名前付き `_1(b)` = ルール内で同一種を強制。
- `get_model(is_netfree=True)` でルールを保持し、`expand(seeds, max_iter, {種: 上限})` で具体反応網へ展開。
- 実モデル例は o10（EGFR シグナル）を参照。
