**World** が初期状態（どの分子が何個・どこに・どの体積で）を保持する。ソルバごとに World 型がある。

```python
import warnings; warnings.filterwarnings('ignore')
from ecell4.prelude import *

edge_lengths = Real3(1, 2, 3)
w1 = gillespie.World(edge_lengths)
w2 = ode.World(edge_lengths)
w5 = meso.World(edge_lengths)
print('edge lengths / volume:', tuple(w1.edge_lengths()), w1.volume())

# 分子を追加（Species は radius, D を持てる）
sp1 = Species('A', 0.0025, 1)
w1.add_molecules(sp1, 10)
print('A count:', w1.num_molecules(Species('A')))

# モデルを bind すると、属性が引き継がれる
m = NetworkModel(); m.add_species_attribute(Species('A', 0.0025, 1))
w1.bind_to(m)
w1.add_molecules(Species('B'), 20)
print('after bind: A=%d B=%d' % (w1.num_molecules(Species('A')), w1.num_molecules(Species('B'))))
```

**要点**: `World(edge_lengths)` で体積を決め、`add_molecules(species, n)` で初期分子を置く（空間 World なら
第3引数で区画/位置指定も可）。`bind_to(model)` でモデルの種属性（radius, D）を World に結びつける。
`run_simulation` の `y0=`/`volume=` は、この World 準備を裏で自動化したもの。
